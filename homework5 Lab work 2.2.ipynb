### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
numbers = input()
lst = numbers.replace('.', '').split()
print(lst)
############
max=lst[0] #максимальне призначаємо першому елементу списку з індексом 0

for i in range(0,len(lst)): #перебираємо список від індексу 0 до довжини списку
    if max<lst[i]: #порівнюємо максимальне значення з і тим елементом
        max=lst[i]

print(f'максимальне значення', max)
#############
min=lst[0] #мінімальне призначаємо першому елементу списку з індексом 0

for i in lst: #перебираємо список від індексу 0 до довжини списку
    if min>i: #порівнюємо мінімальне значення з і тим елементом
        min=i

print(f'мінімальне значення', min)
```

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

#median-lesson4  
#avg-lesson3
```

```python
#рохраховує мінімальне та максимальне значення
max=list_with_random_numbers[0]
min=list_with_random_numbers[0]

for i in range (0, len(list_with_random_numbers)):
    
    if max<list_with_random_numbers[i]:
     max=list_with_random_numbers[i]
    if min>list_with_random_numbers[i]:
        min=list_with_random_numbers[i]
        
print('мінімальний елемент', min)
print('максимальний елемент', max)
```

```python
#середнє значення послідовності
lst = list_with_random_numbers
suma = 0
counter = 0
avg_counter = None


for i in lst:
    counter += 1
    suma += i
    
avg_counter = suma / counter
print(avg_counter)
```

```python
#медіану послідовності
lst = list_with_random_numbers
lst.sort()
if len(lst) % 2 == 0:
    print((lst[len(lst) // 2 - 1] + lst[len(lst) // 2]) / 2)
else:
    print(lst[len(lst) // 2])
    
#import numpy as np
#np.median(lst)
```
