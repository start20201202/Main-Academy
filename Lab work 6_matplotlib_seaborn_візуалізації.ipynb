```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
a=df['a'].values.tolist()
b=df['b'].values.tolist()
c=df['c'].values.tolist()
```

```python
# Create a 2x2 grid of subplots and set the figure size
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Generate sample data
#пряма пропорційність
x = [1, 2, 3]
y = [2, 4, 6]

# Plot the data in each subplot
ax[0, 0].plot(a, 'b-', label='a')
ax[0, 1].plot(b, 'b-', label='b')
ax[1, 0].plot(c, 'b-', label='c')
ax[1, 1].plot(x, y, 'b-', label='xy')

# Add a title for the entire figure
fig.suptitle('diagrams')

# Display the figure
plt.show()
```

```python
#Завантажуємо файл
#  https://docs.google.com/spreadsheets/d/1oJ2fKDqy2SNaygTf_ow5sXjmm0Mi4GkUEeD0U7SgV-I/edit?usp=sharing
sheet_id='1oJ2fKDqy2SNaygTf_ow5sXjmm0Mi4GkUEeD0U7SgV-I'
sheet_name = 'Orders'
df=pd.read_csv(f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}")
df
```

```python
city = df.groupby(['City']).agg({'City': 'count'}).rename(columns={'City':'count_all'}).\
sort_values(by=['count_all'], ascending=False).reset_index()[:10]
```

```python
city.head()
```

```python
#a=city['City'].values.tolist()
#b=city['count_all'].values.tolist()

plt.figure(figsize=(5, 5))
plt.title(r'Sales by City', fontsize=12)
plt.pie(city.count_all, labels=city.City, autopct='%1.1f%%')
plt.show()
```

```python
vals = city.count_all
labels = city.City

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, wedgeprops=dict(width=0.5))
```

```python
customer = df.groupby(['CustomerName']).agg({'CustomerName': 'count'}).rename(columns={'CustomerName':'count_all'}).\
sort_values(by=['count_all'], ascending=False).reset_index()[:10]
```

```python
customer.head()
```

```python
a=customer['CustomerName'].values.tolist()
b=customer['count_all'].values.tolist()
```

```python
plt.figure(figsize=(11, 5))

plt.bar(a, b)
```

```python
plt.title(r'Sales by Customers', fontsize=12)
plt.barh(a, b)
```
