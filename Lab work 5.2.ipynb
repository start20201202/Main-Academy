Створіть нову колонку "Actual price", яка утворюється із "Purchase Price" та націнки за принципом:
- якщо "Purchase Price" > 50, націнка 20%
- якщо "Purchase Price" > 100, націнка 10%
- в інших випадках націнка 30%
   

```python
#lst = []
#for item in df.purchase_price:
#    if item > 50:
#        lst.append('20%')
#    elif item > 100:
#        lst.append('10%')     
#    else:
#        lst.append('30%')
        
#df['actual_price'] = lst

def calc_alctual_price(data):
    price = data['purchase_price']
    if price > 100:
        return price * 1.1
    elif price > 50:
        return price * 1.2
    else:
        return price * 1.3
    
df['actual_price'] = df.apply(calc_alctual_price, axis=1)
```
