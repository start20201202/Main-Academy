```python
import requests
import json
import pandas as pd
```

```python
start = '20220101' 
end = '20221231'

url = f'https://bank.gov.ua/NBU_Exchange/exchange_site?start={start}&end=%20{end}&valcode=usd&sort=exchangedate&order=desc&json'

req = requests.get(url)

```

```python
req
```

```python
src = req.json()
```

2. Отримати курс долара за останній рік.

```python
src
```

3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
dct = {
    'data': [],
    'rate': []
}

for i in range(len(src)):
    dct['data'].append(src[i]['exchangedate'])
    dct['rate'].append(src[i]['rate'])
```

```python
df = pd.DataFrame(dct)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.data = pd.to_datetime(df.data, format='%d.%m.%Y')
```

```python
df.dtypes
```

```python
df['month_year'] = df.data.dt.strftime('%m-%Y')
```

```python
df.head()
```

```python
#Вивести середнє значення та відхилення курсу за кожний місяць.
df_rate = df.groupby('month_year').agg({'rate': ['mean','std']}).reset_index().reset_index() #середнє значення
```

```python
df_rate
```

13. Підрахуйте кількість замовлень для кожного клієнта використовуючи підзапити

```python
result = pd.read_sql_query ( 
'select c.cust_id, s.count_orders from Customers as c \
join\
(select o.cust_id,count(o.order_num) as count_orders from Orders as o group by o.cust_id) as s \
on c.cust_id=s.cust_id ', con)
result
```

14. Виведіть список клієнтів та їх замовлення (використайте JOIN)

```python
result = pd.read_sql_query( 
'select c.cust_name,o.order_num,o.order_date from Customers as c join Orders as o on c.cust_id=o.cust_id \
order by c.cust_id asc, o.order_num asc', con)
result
```

15. Зробіть два запити: список товарів, ціна яких менша 5 та список товарів, ціна який більша рівна 5. Використайте UNION для об'єднання цих запитів

```python
result = pd.read_sql_query( 
'select p.prod_name, "менше 5" as status from Products as p where p.prod_price < "5" \
union all select p.prod_name, "більше рівно 5" as status from Products as p where p.prod_price >= "5"', con)
result
```
