```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
a=df['a'].values.tolist()
b=df['b'].values.tolist()
c=df['c'].values.tolist()
```

```python
# Create a 2x2 grid of subplots and set the figure size
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Generate sample data
#пряма пропорційність
x = [1, 2, 3]
y = [2, 4, 6]

# Plot the data in each subplot
ax[0, 0].plot(a, 'b-', label='a')
ax[0, 1].plot(b, 'b-', label='b')
ax[1, 0].plot(c, 'b-', label='c')
ax[1, 1].plot(x, y, 'b-', label='xy')

# Add a title for the entire figure
fig.suptitle('diagrams')

# Display the figure
plt.show()
```

Створити візуалізацію, аналогічно рисунку

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
plt.figure(figsize=(11, 5))

plt.plot(a, 'r-',linestyle='dashed', label='a')
plt.plot(b, color='orange',linewidth=10, label='b')
plt.plot(c, 'y-',linewidth=1,marker="o", label='b')

# Add legend 
plt.legend(bbox_to_anchor =(0.65, 1.25))

plt.show()
```
